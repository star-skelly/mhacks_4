```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
song_data = pd.read_csv("songs_data.csv")
df = pd.DataFrame(song_data, columns = ["Track Name", "Artist Name(s)", "Danceability","Energy","Key","Loudness","Mode","Speechiness","Acousticness","Instrumentalness","Liveness","Valence","Tempo","Time Signature"])
df.head()
```

```python
basic = pd.DataFrame(song_data, columns = ["Track Name", "Artist Name(s)","Danceability","Energy","Loudness", "Acousticness"])
```

```python
plt.scatter(df["Danceability"], df["Energy"], c=df["Loudness"])
```

```python
fig = plt.figure()
ax2 = fig.add_subplot( projection='3d')

ax2.scatter(basic["Danceability"], basic["Energy"], basic["Loudness"], c=basic["Acousticness"])

ax2.set_xlabel('Danceability')
ax2.set_ylabel('Energy')
ax2.set_zlabel('Loudness')
plt.show()
```
