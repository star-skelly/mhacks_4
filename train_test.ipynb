```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
song_data = pd.read_csv("songs_data.csv")
df = pd.DataFrame(song_data, columns = ["Track Name", "Artist Name(s)", "Danceability","Energy","Loudness","Speechiness","Acousticness","Instrumentalness","Liveness","Tempo", "Valence"])
df.head()
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
#normalize the data
df_ints = pd.DataFrame(song_data, columns = ["Danceability","Energy","Loudness","Speechiness","Acousticness","Instrumentalness","Liveness","Tempo", "Valence"])
normed=(df_ints-df_ints.mean())/df_ints.std()
normed=(df_ints-df_ints.min())/(df_ints.max()-df_ints.min())
normed.head()
```

```python
#clean up the data
for column in ["Danceability","Energy","Loudness","Speechiness","Acousticness","Instrumentalness","Liveness","Tempo", "Valence"]:
    mean = float(normed[column].mean(skipna = True))
    normed[column] = normed[column].replace(np.NaN, mean)
```

```python
train_data = normed.iloc[5:,]
train_data.head()
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(train_data)
```

```python
song_i = 1
test = normed.iloc[song_i:song_i+1:,]
test.head()
```

```python
#find the nearest neighbors
distances, indices = knn.kneighbors(test, n_neighbors=3)
print("d", distances)
print("i", indices)
```

```python
print(df["Track Name"].iloc[song_i], "by", df["Artist Name(s)"].iloc[song_i])
```
